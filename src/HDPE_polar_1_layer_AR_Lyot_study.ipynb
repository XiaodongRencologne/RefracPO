```python
from twolenspy.HDPE_lens import HDEP_2lens
import h5py as h5
import numpy as np
import torch as T
import matplotlib.pyplot as plt
```

```python
PO_sampling_file = 'twolenspy/HDPE_PO_points.h5'
```

## Using uniform sampling to study the polarizaiton rotation at 45deg

```python
dx_list = [0]#[20,40,60,80,100,120]
angle_list = [45]#[0,15,30,45,60,75,90,105,120,135,150,165]
freq_list = [120]#[115,120,130,140,150,160,170,180]#[70,75,80,85,90,95,100,105,110]
for freq in freq_list:
    factor = 1
    Field_factor = 90/freq
    if freq <=90:
        freq_check = 90
    elif freq <= 120:
        freq_check = 120
    elif freq <= 150:
        freq_check = 150
    elif freq >150:
        freq_check = 150
        factor = 2
    for angle in angle_list:
        for item in dx_list:
            print('#------------------')
            print(freq, angle,item)
            with h5.File(PO_sampling_file,'r') as f:
                PO_no = f[str(freq_check)+'GHz'][str(item)][str(45)]['POpoints'][:]
                print(PO_no[4])
            Model1 = HDEP_2lens(freq,
                        feedpos = [item,0,0],
                        feedrot = [0,0,angle],
                        AR_file = 'AR_LIB/HDPE_ARC2.h5',
                        groupname= str(freq)+'GHz',
                        outputfolder = '../Data/HDPE/AR/'+str(freq)+'GHz/',
                        grid_sizex = 0.2*Field_factor,
                        grid_sizey = 0.2*Field_factor)
            
            Model1.run_po(np.array(PO_no[0:2],dtype =np.int32)*factor ,
                          np.array(PO_no[2:4],dtype =np.int32)*factor ,
                          Lyot_N=np.array([PO_no[4][0]*factor ,PO_no[4][1]*factor ],dtype =np.int32),
                         sampling_type = 'less',device = T.device('cuda'))

            Model1.plot_beam(cmap = 'viridis',plot = False)
# 120GHz, 0mm, 135 missingARfile
```
