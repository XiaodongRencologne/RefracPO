```python
import numpy as np
import matplotlib.pyplot as plt
from Refracpo.FresnelCoeff_0 import read_Fresnel_coeffi_AR as read_AR
from Refracpo.antiReflection import read_Fresnel_coeffi_AR as read_AR_v2
```

```python
filename = 'AR_LIB/HDPE_ARC2.h5'
```

```python
F_1, F_2 = read_AR(filename, '120GHz', 1, 1.525)
F_1_v2, F_2_v2 = read_AR_v2(filename,  '120GHz', 1, 1.525)
```

```python
theta = np.linspace(0, 90, 100)*np.pi/180
```

```python
t_p1, t_s1, r_p1, r_s1 = F_1(theta)
t_p2, t_s2, r_p2, r_s2 = F_1_v2(theta)

dt  = t_p1 - t_p2
ds  = t_s1 - t_s2
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1)
ax[0].plot(theta*180/np.pi, dt.real, label='t_p1 - t_p2')
ax[0].plot(theta*180/np.pi, ds.real, label='t_s1 - t_s2')

ax[1].plot(theta*180/np.pi, dt.imag, label='t_p1 - t_p2')
ax[1].plot(theta*180/np.pi, ds.imag, label='t_s1 - t_s2')
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1)
ax[0].plot(theta*180/np.pi, t_p1.real, label='t_p1 - t_p2')
ax[0].plot(theta*180/np.pi, t_p2.real, label='t_s1 - t_s2')

ax[1].plot(theta*180/np.pi, t_p1.imag, label='t_p1 - t_p2')
ax[1].plot(theta*180/np.pi, t_p2.imag, label='t_s1 - t_s2')
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1)
ax[0].plot(theta*180/np.pi, np.abs(t_p1), label='t_p1 - t_p2')
ax[0].plot(theta*180/np.pi, np.abs(t_p2), label='t_s1 - t_s2')

ax[1].plot(theta*180/np.pi, np.angle(t_p1), label='t_p1 - t_p2')
ax[1].plot(theta*180/np.pi, np.angle(t_p2), label='t_s1 - t_s2')
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1)
ax[0].plot(theta*180/np.pi, np.abs(t_p1)-np.abs(t_p2), label='t_p1 - t_p2')
ax[0].plot(theta*180/np.pi, np.abs(t_s1)-np.abs(t_s2), label='t_s1 - t_s2')

ax[1].plot(theta*180/np.pi, np.angle(t_p1)-np.angle(t_p2), label='t_p1 - t_p2')
ax[1].plot(theta*180/np.pi, np.angle(t_s1)-np.angle(t_s2), label='t_s1 - t_s2')
```
